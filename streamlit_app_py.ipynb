```
import streamlit as st
import pandas as pd
import time
import random
from datetime import datetime

# --- CONFIGURATION & STYLING ---
st.set_page_config(
    page_title="Aether-Link Control Plane",
    page_icon="⚡",
    layout="wide",
    initial_sidebar_state="expanded",
)

# Custom CSS for the "LXD-Dark" aesthetic
st.markdown("""
    <style>
    .main { background-color: #0e1117; }
    .stMetric { background-color: #161b22; border: 1px solid #30363d; padding: 15px; border-radius: 10px; }
    .stActionButton { background-color: #ff4d00; color: white; }
    .terminal-text { font-family: 'Fira Code', monospace; color: #39ff14; font-size: 12px; }
    </style>
    """, unsafe_allow_html=True)

# --- SESSION STATE INITIALIZATION ---
if 'bot_active' not in st.session_state:
    st.session_state.bot_active = True
if 'logs' not in st.session_state:
    st.session_state.logs = [
        f"[{datetime.now().strftime('%H:%M:%S')}] Aether-Link Engine Initialized.",
        f"[{datetime.now().strftime('%H:%M:%S')}] Infura Bridge connected via da16bf...32e2",
    ]

# --- SIDEBAR: INFRASTRUCTURE ---
with st.sidebar:
    st.title("🛡️ Aether-Link")
    st.subheader("Compute & Cloud")

    st.info("**Environment:** `Rust-conda` (main)")
    st.success("**LXD Node:** `vps-node-01` Online")

    st.divider()

    st.subheader("Quick Controls")
    if st.button("Restart Telegram Bot", use_container_width=True):
        st.session_state.logs.append(f"[{datetime.now().strftime('%H:%M:%S')}] Bot process restarted...")

    if st.button("Flush Infura Cache", use_container_width=True):
        st.session_state.logs.append(f"[{datetime.now().strftime('%H:%M:%S')}] RPC cache cleared.")

# --- MAIN DASHBOARD ---
st.header("Cloud Orchestrator: System Dashboard")

# Top Metrics Row
col1, col2, col3, col4 = st.columns(4)
with col1:
    st.metric("Arbitrage Spread", "0.42%", "+0.03%", delta_color="normal")
with col2:
    st.metric("Infura Latency", "14ms", "-2ms", delta_color="inverse")
with col3:
    st.metric("Active Skills", "12", "Ready")
with col4:
    st.metric("Gas Price", "18 Gwei", "-5 Gwei", delta_color="inverse")

st.divider()

tab1, tab2, tab3 = st.tabs(["🚀 Control Plane", "🔗 Bridge Intelligence", "📄 Documentation"])

with tab1:
    c1, c2 = st.columns([2, 1])

    with c1:
        st.subheader("Active Instances (LXD/Docker)")
        instances = pd.DataFrame({
            "Instance Name": ["hb-client-primary", "hb-gateway-01", "mcp-intel-node", "tg-bridge"],
            "Status": ["✅ Running", "✅ Running", "⏳ Starting", "✅ Running"],
            "IP Address": ["10.155.42.102", "10.155.42.105", "10.155.42.110", "10.155.42.112"],
            "CPU": ["12%", "4%", "22%", "1%"]
        })
        st.table(instances)

        st.subheader("Market Spread Visualization")
        chart_data = pd.DataFrame(
            [0.35, 0.38, 0.34, 0.42, 0.45, 0.48, 0.41, 0.52, 0.48, 0.42],
            columns=['Spread %']
        )
        st.line_chart(chart_data)

    with c2:
        st.subheader("System Terminal")
        log_container = st.container(height=400)
        for log in reversed(st.session_state.logs):
            log_container.markdown(f"<p class='terminal-text' style='margin:0;'>&gt; {log}</p>", unsafe_allow_html=True)

        if st.toggle("Auto-Refresh Terminal", value=True):
            time.sleep(1)
            # Simulating live data injection
            if random.random() > 0.7:
                st.session_state.logs.append(f"[{datetime.now().strftime('%H:%M:%S')}] Spread Check: {random.uniform(0.3, 0.6):.2f}%")
                st.rerun()

with tab2:
    st.subheader("Infura RPC Bridge Configuration")
    st.code("""
# eth-json-rpc-infura Middleware State
API_KEY = "da16bf38c04a4ec084f2d978120e32e2"
NETWORK = "mainnet"
ENDPOINT = f"https://{NETWORK}.infura.io/v3/{API_KEY}"
    """, language="python")

    st.write("### Bridge Health")
    st.progress(85, text="API Quota Used (85%)")
    st.warning("Approaching free-tier limit. Consider rotating API keys.")

with tab3:
    st.markdown("""
    ### Aether-Link Setup
    1. **Environment**: Ensure `conda activate rust-env` is active.
    2. **Infura**: Uses the `json-rpc-engine` middleware via Node bridge.
    3. **Telegram**: Bot active at `/workspaces/Rust-conda/Condor_Hummingbot_Feeder.ipynb`.

    **Support commands:**
    - `/status` : Check engine health
    - `/portfolio` : Current balances across Binance/KuCoin
    """)

# Footer
st.divider()
st.caption(f"Aether-Link Orchestrator | Engine: Stable | Last update: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
